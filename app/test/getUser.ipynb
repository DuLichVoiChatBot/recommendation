```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore
import requests
import json
from firebaseHelper import getCollection
```

```python
docs = getCollection('users')
```

```python
docs = list(docs)
docs = [doc.to_dict() for doc in docs]
payload = {}
headers = {}
```

```python
for doc in docs:
    doc['interest'] = [str(i) for i in doc['interest']]
    url = "http://127.0.0.1:8000/items/" + '&'.join(doc['interest'])
    response = requests.request("GET", url, headers=headers, data=payload)
    rp = json.loads(response.text)
    interests = []
    interests.extend(rp)
    doc['interest'] = interests
```
